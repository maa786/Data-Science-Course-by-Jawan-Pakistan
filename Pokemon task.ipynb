###### here i can check the percentile of hp columns

```python
hp_seventy_five_percitle_data = data["HP"].quantile(0.75)
hp_seventy_five_percitle_data
```

```python
hp_fifty_percitle_data = data["HP"].quantile(0.50)
hp_fifty_percitle_data
```

```python
hp_25_percentile_data = data["HP"].quantile(0.25)
hp_25_percentile_data
```

```python
def add_hp(val):
    if val >= hp_seventy_five_percitle_data:
        return "High Hp"
    elif val >= hp_fifty_percitle_data and val < hp_seventy_five_percitle_data:
        return "Medium Hp"
    else:
        return "Low Hp"
```

```python
data["HP_High_Low"] = data["HP"].apply(add_hp)
```

```python
data["HP_High_Low"].value_counts()
```

### Here the second column i can Add_Speed_Column according to the speed column and apply the function

```python
#heres the quick review about the speed column
data["Speed"].describe()
```

```python
# here i can find values using the percentile function

seventy_5_percentile , fifty_percentile , twenty_5_percentile = data["Speed"].quantile([0.75,0.50,0.25])
seventy_5_percentile , fifty_percentile , twenty_5_percentile
```

```python
#define the function

def set_speed(num):
    if num >= seventy_5_percentile:
        return "high speed"
    elif num >= fifty_percentile and num < seventy_5_percentile:
        return 'Medium Speed'
    else:
        return "Low Speed"
```

```python
#here i can add column and apply function

data["Add_Speed_Column"] = data["Speed"].apply(set_speed)
```
